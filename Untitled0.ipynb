```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
data = pd.read_csv(io.BytesIO(uploaded['data.csv']))
```

```python
cleaned_data = data.dropna()
```

```python
print(cleaned_data.isnull().sum())
```

```python
descriptive_stats = cleaned_data[['employees_count', 'price (USD)']].describe()
print(descriptive_stats)
```

```python
correlation_matrix = cleaned_data[['employees_count', 'price (USD)']].corr()
print(correlation_matrix)
```

```python
plt.scatter(cleaned_data['employees_count'], cleaned_data['price (USD)'])
plt.xlabel('Employees Count')
plt.ylabel('Price (USD)')
plt.title('Scatter Plot of Employees Count vs Price (USD)')
plt.show()
```

```python
X = cleaned_data['employees_count']
y = cleaned_data['price (USD)']
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X = cleaned_data[['employees_count']]

X = sm.add_constant(X)

vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

print(vif_data)
```

```python
missing_data = data.isnull().sum()
print(missing_data)


data['Symbol'].fillna('Unknown', inplace=True)
data['country'].fillna('Unknown', inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


country_employee_count = data.groupby('country')['employees_count'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,8))
country_employee_count.plot(kind='bar')
plt.title('Toplam alan Saysna Gre lkeler')
plt.xlabel('lkeler')
plt.ylabel('alan Says')
plt.show()

plt.figure(figsize=(10,6))
sns.histplot(data['price (USD)'], kde=True)
plt.title('irketlerin Hisse Fiyatlar Dalm')
plt.xlabel('Hisse Fiyat (USD)')
plt.ylabel('Frekans')
plt.show()
```

```python
data['average_employment'] = data['employees_count'] / 12
data[['Name', 'average_employment']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='employees_count', y='price (USD)')
plt.title('alan Says ve Hisse Fiyat Arasndaki liki')
plt.xlabel('alan Says')
plt.ylabel('Hisse Fiyat (USD)')
plt.show()


plt.figure(figsize=(12,8))
sns.boxplot(x='country', y='employees_count', data=data)
plt.title('lkelere Gre alan Says Dalm')
plt.xticks(rotation=90)
plt.xlabel('lkeler')
plt.ylabel('alan Says')
plt.show()


numerical_cols = data[['employees_count', 'price (USD)', 'average_employment']]
correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Korelasyon Matrisi')
plt.show()
```
